```python
import zipfile
import pandas as pd
 
data = pd.read_csv('archive.zip',compression='zip')
data = data.head(5)
y = data['sentiment'].replace({'positive': 1, 'negative': 0})
print(y, data)
```

```python
inputMeans=[]
for index, row in data.iterrows():
    inputMeans.append([])
    rowText = nlp(row['review'])
    for token in rowText:
        lineToken= nlp(token.text)
        vectorTokenized = lineToken.vector
        meanOfVector = vectorTokenized.mean()
        inputMeans[index].append(meanOfVector)
```
